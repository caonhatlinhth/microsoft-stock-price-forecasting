```python
import pandas as pd
```

```python
microsoft_stock_data = '/content/drive/My Drive/Colab Notebooks/Time Series/Data/microsoft_stock_price.csv'
```

```python
df = pd.read_csv(microsoft_stock_data)
df.head()
```

```python
# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract date and time components
df['date'] = [d.date() for d in df['Date']]
df['time'] = [d.time() for d in df['Date']]

# Drop the original 'Date' column
df.drop(['Date'], axis=1, inplace=True)

cols = ['date', 'time'] + [col for col in df if col not in ['date', 'time']]
df = df[cols]

# Print the head to check the result
print(df.head())
```

```python
df.to_csv('/content/drive/My Drive/Colab Notebooks/Time Series/ms_stock_new.csv', index=False)
```
